```python
def check_tags_from_dict(tags, word_list):
    word_present = 0
    
    # Tags can be quite ugly so must format clean into separate words to make them usable for the set intersection
    tags = tags.lower()
    tags = tags.replace(';',' ').replace(',',' ').replace('(',' ').replace(')',' ').replace('&',' ').replace('|',' ').split()
    
    # Make sets from the list of words
    x = set(tags)
    y = set(word_list)
    
    # Use intersection function of set type to quickly find which word are in both lists/sets
    z = x.intersection(y)
    
    # If the intersection is larger than 0, then there is at least a common word between both sets 
    if len(z) > 0:
        word_present = 1
    return word_present
```
